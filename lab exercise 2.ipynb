```python
class Solution:
    def myAtoi(self, s: str) -> int:
        s =s.strip()
        if  len(s)== 0:  #to check is string is empty
            return 0
        if s[0] == "-":
            sign = -1
            s = s[1:]
        elif s[0] == "+":
            sign = 1
            s = s[1:]
        else:
            sign = 1
        num = 0
        for i in s:
            if i not in "0123456789":
                break
            num = num * 10 + int(i)
        return num * sign
```

```python
#13. Write a program to accept string/sentences from the user till the user enters “END” to. Save
# the data in a text file and then display only those sentences which begin with an uppercase
# alphabet.

file_name = input("enter the file name")

f = open(file_name, "a")
while True:
    k = input("enter the text to append")
    if k == "END":
        break
    else:
        f.write(k + "\n")
f.close()
f = open(file_name, "r")

for i in f:
    if i[0].isupper():
        print(i)
f.close()


```

```python
# 14. Write a program to enter the following records in a binary file:
# Item No integer
# Item_Name string
# Qty integer
# Price float
# Number of records to be entered should be accepted from the user. Read the file to display the
# records in the following format:
# Item No:
# Item Name :
# Quantity:
# Price per item:
# Amount: ( to be calculated as Price * Qty)

import pickle
file = open("data.dat", "wb")
n = int(input("enter the value"))
for i in range(n):
    data = {}
    data["item_no"]= int(input("enter the no of item"))
    data["item_name"]= input("enter the name")
    data["qty"] = int(input("enter the no of qty"))
    data["price"] = int(input("enter the price"))
    pickle.dump(data, file)

file.close()

file = open("data.dat", "rb")
while True:
    try:
        data = pickle.load(file)
        print("Item No:", data["item_no"])
        print("Item Name:", data["item_name"])
        print("Quantity:", data["qty"])
        print("Price per item:", data["price"])
        print("Amount:", data["price"] * data["qty"])
    except EOFError:
        break
file.close()
```

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        string = ""
        for i in range(len(strs)):
            for j in range(len(strs[i])):
               if strs[i][j] == strs[i+1][j]:
                    string += j
                    return string
                
```
