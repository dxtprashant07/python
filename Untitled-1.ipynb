```python
#decorator
# @decorator wraps the add function inside wrapper.

# Before calling add, it logs "Calling function...".

# After calling add, it logs "Function finished".

# The return value from add is preserved.

from functools import wraps
def decorator(factrec):
    @wraps(factrec)
    def wrapper(factrec):
        def inner(*args, **kwargs):
            print("this is before function execution")
            result = factrec(*args, **kwargs)
            print("this is after function execution")
            return result
        return inner
    return wrapper


@decorator
def factrec(n):

    if n == 1 or n==0:
        return 1
    else:
        
        return n*factrec(n-1)
      
n =int(input())
k = factrec(n)
print(k)

```

```python
def decor(func):
    @wraps(func)
    def wrapper(check_positive):
        if check_positive < 0:
            raise ValueError("Input must be a positive integer")
        
    
        return func(check_positive)
    return wrapper
@decor
def func(check_positive):
    return check_positive > 0
n = int(input())
result = func(n)
print(result)
```

```python
def greet(func):

    def wrapper(*args, **kwargs):
        
        print("Hello!")
        func(*args, **kwargs)
        print("Goodbye!")
    return wrapper

@greet
def say_name(name):
    print(f"My name is {name}.")

@greet
def add(a,b):
    print(f"The sum is {a+b}")

say_name("Alice")

add(3,4)
```

```python
def outer_function(addition):
    def wrpper(*args, **kwargs):
        print("this is before function execution")
        result = addition(*args, **kwargs)
        num3 = float(input("enter a number"))
        result += num3
        print("this is after function execution")
        return result
    return wrpper

@outer_function
def addition(num1,num2):
    return num1 + num2
num1 = float(input("enter a number"))
num2 = float(input("enter a number"))
result = addition(num1,num2)
print(result)
```

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
for number in count_up_to(5):
    print(number)
```
