```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y)
plt.title("line plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(101)
data =np.random.normal(50,10,200)
plt.boxplot(data)
plt.title("boxplot")
plt.ylabel("Values")
plt.show()
```
