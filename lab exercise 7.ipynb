```python
# 1. Define a class to represent a bank account. Include the following
# members:Â 
# Data Members:
# a.Name of the Depositor
# b.Account Number
# c.Type of Account
# d.Balance amount in the account
# Data Methods:
# a.To assign initial values
# b.To deposit an amount
# c.To withdraw an amount
# d.To display name and balance.

# Output:
# 1. New customer
# 2. Deposit
# 3. Withdrawal
# 4. Display
# 5. Exit




class BankAccount():
    def __init__(self, name, account_number, account_type, balance = 0):
        self.name = name
        self.account_number = account_number
        self.account_type = account_type
        self.balance =balance

    def initial_value(self):

        self.name = input("enter your name:")
        self.account_number = int(input("enter your account number:"))
        self.account_type = input("enter your account type:")
        self.balance = int(input("enter your balance:"))
        print(f"Account created for {self.name}.")
    
    def deposit(self):
        amount = int(input("enter your deposit amount:"))
        self.balance = self.balance + amount
        
        print(f"deposit money:{self.balance}")
    def withdraw(self):
        amount = int(input("enter your withdraw amount:"))
        if self.balance>amount:
            self.balance-=amount
            print(f"deposit money:{self.balance}")
        else:
            print(f"not enough money ,deposit money")
    def display(self):
        print(f"Account created for {self.name}.")
        print(f"Account Number: {self.account_number}")
        print(f"account type: {self.account_type}")
        print(f"balance:{self.balance}")


m = BankAccount("","", "")


print("""1. New customer
2. Deposit
3. Withdrawal
4. Display
5. Exit""")



while True:
    new = int(input("enter your choice"))
    if new == 1:
        m.initial_value()
    elif new == 2:
        m.deposit()
    elif new == 3:
        m.withdraw()
    elif new == 4:
        m.display()
    elif new == 5 :
        break
```

```python
class staff:
    def __init__(self, name, code):
        self.name = name
        self.code = code
    def display(self):
        print(f"Name: {self.name}")
        print(f"Code: {self.code}")
        
class Teacher(staff):
    def __init__(self, name, code, subject, publication):
        super().__init__(name, code)
        self.subject = subject
        self.publication = publication
    def display(self):
        super().display()
        print(f"Subject: {self.subject}")
        print(f"Publication: {self.publication}")

class typist(staff):
    def __init__(self, name, code ,speed):
        super().__init__(name, code)
        self.speed = speed
    def display(self):
        super().display()
        print(f"Speed: {self.speed}")

class officer(staff):
    def __init__(self, name, code, grade):
        super().__init__(name, code)
        self.grade = grade
    def display(self):
        super().display()
        print(f"Grade: {self.grade}")

class regular(typist):
    def __init__(self, name, code, speed,salary):
        super().__init__(name, code, speed)
        self.salary = salary
    def display(self):
        super().display()
        print(f"Salary: {self.salary}")

class casual(typist):
    def __init__(self, name, code, speed, dailywages):
        super().__init__(name, code, speed)
        self.dailywages = dailywages
    def display(self):
        super().display()
        print(f"Daily Wages: {self.dailywages}")

staff_list = []
while True:
    print("1. Add Teacher")
    print("2. Add Typist")
    print("3. Add Officer")
    print("4. Add Regular Typist")
    print("5. Add Casual Typist")
    print("6. Display All Staff")
    print("7. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        name = input("Enter name: ")
        code = input("Enter code: ")
        subject = input("Enter subject: ")
        publication = input("Enter publication: ")
        staff_list.append(Teacher(name, code, subject, publication))
    elif choice == "2":
        name = input("Enter name: ")
        code = input("Enter code: ")
        speed = input("Enter speed: ")
        staff_list.append(typist(name, code, speed))
    elif choice == "3":
        name = input("Enter name: ")
        code = input("Enter code: ")
        grade = input("Enter grade: ")
        staff_list.append(officer(name, code, grade))
    elif choice == "4":
        name = input("Enter name: ")
        code = input("Enter code: ")
        speed = input("Enter speed: ")
        salary = input("Enter salary: ")
        staff_list.append(regular(name, code, speed, salary))
    elif choice == "5":
        name = input("Enter name: ")
        code = input("Enter code: ")
        speed = input("Enter speed: ")
        dailywages = input("Enter daily wages: ")
        staff_list.append(casual(name, code, speed, dailywages))
    elif choice == "6":
        for staff in staff_list:
            staff.display()
    elif choice == "7":
        break
    else:
        print("Invalid choice. Please try again.")
```
